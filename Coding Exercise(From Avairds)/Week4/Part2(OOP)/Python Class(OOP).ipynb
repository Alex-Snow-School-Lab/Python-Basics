### Encapsulation


```
class Person:
    def __init__(self, name, age=0):
        self.name = name
        self._age = age

    def display(self):
        print(self.name)
        print(self._age)

person = Person('John', 40)
#accessing using class method
person.display()
#accessing directly from outside
print(person.name)
print(person._age)
```

### Inheritance

```
class Person:
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname

  def printname(self):
    print(self.firstname, self.lastname)
    



x = Person("John", "Doe")
x.printname()
```

```
class doctor(Person):
   def __init__(self, fname, lname):
    Person.__init__(self, fname, lname)

 
x = doctor("Mike", "Olsen")
x.printname()
  
```

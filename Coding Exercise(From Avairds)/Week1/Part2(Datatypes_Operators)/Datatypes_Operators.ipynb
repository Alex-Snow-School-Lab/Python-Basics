### **Basic Functions**

`print()` : ရေးထားသော data အကြောင်းကို system output တွင်ထုတ်ပြခြင်း


`type() `: data type/class အမျိုးအစား ထုတ်ကြည့်ခြင်း 


`int(),float(),str()` : type casting (တန်ဖိုးတစ်ခုကို လိုချင်တယ့် datatype သို့ပြောင်းခြင်း)


```
# Print( )
print('Using print()')
print('Hello World')
print(2020)
print('Welcome', 'from', 'AVAIRDS')
print(21, 'June', 2020)
print()
```

```
# Type()
print('Using Type()')
type('Welcome from AVAIRDS')
print('Datatype of 2020 is', type(2020))
print('Datatype of 21.06 is', type(21.06))
print('Datatype of True is', type(True))
```

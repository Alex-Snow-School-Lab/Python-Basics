# List


### Basic List Operations

```
a = [10,20,30,40,50,60,70,80,90]
print(len(a))
print(min(a))
print(max(a))
```

```
print (a + [100,110,120])
```

### List slicing


```
a[1:-3] 
```

```
a[2:-4] = "a" ,"b" ,"c"
a
```

```
#slicing 3 index
a[2:8:3]
```

```
# slicing 2index backward
a[7::-3]
```

### List inside list

```
a[3] = ["e" , "f" , "g" ,"h"]
a
```

```
a[3][-1::-2]
```

### List method

```
c= [4,5,6,6,9,0]
c.count(6)
```

```
c.append([76,78])
c
```

```
c.extend([7,8,9])
c
```

```
c.pop()
```

```
c.insert(2,2)
c
```

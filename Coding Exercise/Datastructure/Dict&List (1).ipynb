# List


### Basic List Operations

```
a = [10,20,30,40,50,60,70,80,90]
print(len(a))
print(min(a))
print(max(a))
```

```
print (a + [100,110,120])
```

### List slicing


```
a[1:-3] 
```

```
a[2:-4] = "a" ,"b" ,"c"
a
```

```
#slicing 3 index
a[2:8:3]
```

```
# slicing 2index backward
a[7::-3]
```

### List inside list

```
a[3] = ["e" , "f" , "g" ,"h"]
a
```

```
a[3][-1::-2]
```

### List method

```
c= [4,5,6,6,9,0]
c.count(6)
```

```
c.append([76,78])
c
```

```
c.extend([7,8,9])
c
```

```
c.pop()
```

```
c.insert(2,2)
c
```

# Dictionary

### Different way of creating dictionary

```
a = {'joe' : 85 , 'peter' : 88 , 'jack' : 90}
a
```

```
b = dict (joe = 90 , peter = 85 , jack = 88)
b
```

```
c = dict ( [  ('joe' , 90)  ,  ('peter' , 85) ,  ('jack' , 95)  ]  )
c
```

### Adding new key and value:

```
 a['kris'] = 85
 a
```

```
del a['kris']
a
```

### Dictionary methods

```
a = {'joe' : 85 , 'peter' : 88 , 'jack' : 90}
b = a.get('peter',0)
print(b)
```

```
b = a.keys()
print(b)
```

```
b = a.values()
print(list(b))
```

### Creating two dimensional dictionary: Method 1: Use list as values


```
marks = {'peter': [80,85,90], 'jose' : [85,90,95]}
print(marks['jose'])
```

### Method 2 : use dictionary as values

```
 marks2= {'jose' : {'maths': 85 , 'english': 90, 'science' : 95} , 'peter' : {'marks': 80 , 'english':85 , 'science': 90}}
 print (marks2['jose']['science'])
```

```
print (marks2['jose']['science'])
```

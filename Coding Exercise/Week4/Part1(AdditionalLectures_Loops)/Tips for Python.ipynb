Tip 1 - Format strings with f-strings

```
name = 'Alex'
my_string = f"Hello {name}"
print(my_string)

i = 10
print(f"{i} sqaured is {i*i}")
```

Tip 2 - Merge two dictionaries

```
d1 = {"name" : 'Alex', 'age' : 25}
d2 = {"name" : 'Alex', 'city' : 'New York'}
merged_dict = {**d1, **d2}
print('Merged dictionaries: ', merged_dict)
```

Tip 3 - Save Memory with Generator Comprehensions

```
import sys

my_list = [i for i in range(100)]
print('Sum of my list:', sum(my_list))
print()
print('Size of my list: ', sys.getsizeof(my_list), 'bytes')

my_gen = (i for i in range(100)
print('Size of my generator: ', sys.getsizeof(my_gen), 'bytes')
```

Tip 4 - Simplify if statement for multiple checks

```
# Using in keyword

c = 'red'
if c == 'red' or c == 'green' or c == 'blue':
  print(c, 'is main color')

colors = ['red', 'yellow', 'white']
if c in colors:
  print(c, 'is main color')
```

**List Comprehension**

Comparing normal for loop and using list comprehension

```
# Exercise - 1
names = ['Johnny', 'Sarah', 'Jack', 'Christopher']

# Normal For Loop
l = []
for person in names:
	l.append(person)
print(l)

# Using List Comprehension
print([person for person in names]) 
```

```
# Exercise - 1.1

# Normal For Loop
l = []
for person in names:
	l.append(person + ' invited me to dinner')
print(l)

# Using List Comprehension
print([person + ' invited me to dinner' for person in names]) 
```

```
# Exercise - 2
movies_and_ratings = {"Interstellar" : 7, 
"Dark Knight" : 8,
"Joker" : 9,
"Titanic" : 10
					}
# Normal For Loop
l = []
for movie in movies_and_ratings:
	if movies_and_ratings[movie] > 8:
		l.append(movie)
      
print(l)

# Using List Comprehension
print([movie for movie in movies_and_ratings if movies_and_ratings[movie] > 8])
```

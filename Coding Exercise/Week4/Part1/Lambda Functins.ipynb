```
# Exercise - 1

# Using normal function
def getSum(num1, num2):
	total = num1 + num2
	return total
print('Using normal function: ', getSum(1, 2))

# Using lambda function
getSum = lambda num1, num2 : num1 + num2
print("Using lambda function: " , getSum(10, 3))
```

```
# Exercise - 2

# Function to compute 3x+1
def compute(x):
	return 3 * x + 1
print('Using normal function: ', compute(1))

ans = lambda x : 3 * x + 1
print("Using lambda function: " , ans(2))
```

```
# Exercise - 3
def build_quadratic_function(a, b, c):
	''' Return the function f(x) = ax^2 + bx +c'''
	return lambda x : a * x ** 2 + b * x + c

f = build_quadratic_function(2, 3, -5)
f(0)

build_quadratic_function(3, 0, 1)(2) # 3x^2 + 1 evaluated for x = 2	
```
